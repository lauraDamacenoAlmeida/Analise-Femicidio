```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = df.drop(['DESCR_COR_VEICULO','PLACA_VEICULO','DESCR_COR_VEICULO','ANO_MODELO','CIDADE_VEICULO','ANO_FABRICACAO','UF_VEICULO','DESCR_MARCA_VEICULO'],axis=1)
```

```python
df.head()
```

```python
df = df.drop(['DATAELABORACAO','BO_EMITIDO','BO_INICIADO','NUMERO_BOLETIM','ANO_BO'], axis=1)
```

```python
df = df.drop(['NUMERO_BOLETIM_PRINCIPAL','LATITUDE','LONGITUDE','RG','RG_UF','DATANASCIMENTO','CORCUTIS','NOMEPESSOA'],axis=1)
```

```python
df.info()
```

```python
df.groupby('PERIODOOCORRENCIA')['PERIODOOCORRENCIA'].count().plot.bar()
plt.title('Total de ocorrências')
```

```python
df.groupby('FLAGRANTE')['FLAGRANTE'].count().plot.bar()
```

```python
madrugada = df.groupby(['FLAGRANTE','PERIODOOCORRENCIA'])['FLAGRANTE'].count().plot.bar()
```

```python
madrugada = df['PERIODOOCORRENCIA']=='DE MADRUGADA'
df[madrugada].groupby('FLAGRANTE')['FLAGRANTE'].count().plot.bar()
```

```python
nacionalidade = df.groupby('NACIONALIDADE')['NACIONALIDADE'].count()
nacionalidade.plot.pie(figsize=(5, 6),autopct='%1.1f%%',shadow=True, labels=None)
plt.legend(labels=nacionalidade.index, loc='upper left') 
```

```python
total = len(df)
brasileira = df[(df['NACIONALIDADE']== 'BRASILEIRA') | (df['NACIONALIDADE']=='BARSILEIRO')]['NACIONALIDADE'].count()
cubanas = df[df['NACIONALIDADE']== 'CUBA']['NACIONALIDADE'].count()
sem_nacionalidade = total-(brasileira+cubanas)
porc_brasileiras = (brasileira* 100)/total
porc_cubanas = (cubanas* 100)/total
porc_sem = (sem_nacionalidade* 100)/total


print('Porcentagem de mulheres Brasileiras que sofreram homicido em 2019:{}%'.format(porc_brasileiras.round(2)))
print('Porcentagem de mulheres Cubanas que sofreram homicido em 2019 :{}% '.format(porc_cubanas.round(2)))
print('Porcentagem de mulheres sem nacionalidade que sofreram homicido em 2019 :{}%'.format(porc_sem.round(2)))
```
